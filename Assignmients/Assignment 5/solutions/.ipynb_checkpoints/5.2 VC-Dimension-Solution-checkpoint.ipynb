This exercise will let you explore the Vapnik-Chervonenkis dimension for logistic regression with polynomial features. We will test the shattering of datasets by fitting a classifier for every possible label combination. The samples will be arranged equally distant from each other on the unit circle.

**Q 5.2.1: You know from the lecture that linear models can shatter sets with 3 samples. Set the regularization penalty for logistic regression to 'l1' with C = 2 and n_samples = 3. What happens? Why?**

** Answer: ** The model can no longer shatter the dataset as the cost for using two weights is now higher than the cost for misclassification. You can see from the plots that the decision boundaries have become axis-aligned, which means one of the weights has become zero. This is the general behavior of L1 regularization as it discourages the use of all weights (also known as sparsity).

**Q 5.2.2: The theoretical VC dimension for polynomial models of degree d with k input variables is $\begin{pmatrix} d + k \\ d \end{pmatrix}$. How many did you shatter? Why is it not the theoretical result?**

**Answer: **The VC dimension is defined as the maximum cardinality of **any** dataset which can be shattered. We chose our samples to be equidistant on the unit circle, which apparently cannot be shattered for the theoretically possible cardinality. This is one of the reasons why the VC dimension is very hard to estimate empirically.
